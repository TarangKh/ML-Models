```python
df = combd_dataframe
```

```python
df.info()
```

```python
df.rename(columns = {"Entity": "Country", "YEar": "Year"}, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = LabelEncoder()
for i in df.columns :
  if df[i].dtype == "object" :
    df[i] = l.fit_transform(df[i])
```

```python
df.shape
```

```python
df.info()
```

```python
## Train and test (Creating model)
```

```python
x = df.drop("Mental disorders", axis = 1)
y = df["Mental disorders"]

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = .20, random_state = 2)
```

```python
print("Printing train test shapes")
print("x train =", xtrain.shape)
print("x test =", xtest.shape)
print("y train =", ytrain.shape)
print("y test =", ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(xtrain, ytrain)

ytrain_pred = lr.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = np.sqrt(mean_squared_error(ytrain, ytrain_pred))

r2 = r2_score(ytrain, ytrain_pred)

print("Linear Regression model for training data")
print("MSE:", mse)
print("RMSE:", rmse)
print("r2 score:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)

ytrain_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = np.sqrt(mean_squared_error(ytrain, ytrain_pred))

r2 = r2_score(ytrain, ytrain_pred)

print("Linear Regression model for training data")
print("MSE:", mse)
print("RMSE:", rmse)
print("r2 score:", r2)
```

```python
# test

ytest_pred = lr.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = np.sqrt(mean_squared_error(ytest, ytest_pred))

r2 = r2_score(ytest, ytest_pred)

print("Linear Regression model for testing data")
print("MSE:", mse)
print("RMSE:", rmse)
print("r2 score:", r2)

ytest_pred = rf.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = np.sqrt(mean_squared_error(ytest, ytest_pred))

r2 = r2_score(ytest, ytest_pred)
print()
print("Linear Regression model for testing data")
print("MSE:", mse)
print("RMSE:", rmse)
print("r2 score:", r2)
```
